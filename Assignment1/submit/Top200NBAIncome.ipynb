### Step 2:
#### Follow the instruction on the website to get the Json format dataset

```python
import http.client, urllib.request, urllib.parse, urllib.error, base64

headers = {
    # Request headers
    'Ocp-Apim-Subscription-Key': '867f36937b164583936c068833e3012a',
}

params = urllib.parse.urlencode({
})

try:
    conn = http.client.HTTPSConnection('api.fantasydata.net')
    conn.request("GET", "/v3/nba/stats/JSON/PlayerSeasonStats/2017?%s" % params, "{body}", headers)
    response = conn.getresponse()
    data = response.read()
    print(data)
    conn.close()
except Exception as e:
    print("[Errno {0}] {1}".format(e.errno, e.strerror))
```

### Step 3:
#### Use json.loads to transfor json format into python dic

```python
dic=json.loads(data)
dic
```

### Step 4:
#### Create lists to instore data and use loop to pick up the data we want which include PlayerID,Name,Team,Position and Games.Use pandas to store the data in the form of dataframe.

```python
import pandas as pd
playerID=[]
name=[]
team=[]
position=[]
game=[]
i=0
while i <= len(dic)-1:
    temp=dic[i]['PlayerID']
    playerID.append(temp)
    temp1=dic[i]['Name']
    name.append(temp1)
    temp2=dic[i]['Team']
    team.append(temp2)
    temp3=dic[i]['Position']
    position.append(temp3)
    temp4=dic[i]['Games']
    game.append(temp4)
    i +=1
player_df=pd.DataFrame({"PlayerID":playerID,"Name":name,"Team":team,"Position":position,"Games":game})
player_df
```

## Webscarpy part:
#### Following codes are are about getting dataset from ESPN by scrapying
### Step 1:
#### Find the all webpage for scraping data 

```python
from requests import get
address=[]
i=1
while i<=12:
    url='http://www.espn.com/nba/salaries/_/year/2017/page/{}'.format(i)
    address.append(url)
    i+=1
address
```

### Step 2:
#### Use BeautifulSoup to get the html

```python
from bs4 import BeautifulSoup as bp # importing Beautiful Soup
rank=[]
i = 0
while i<=11:
    response=get(address[i])
    html_soup=bp(response.text,'html.parser')# python's in built library HTML parser
    rank.append(html_soup.find_all('tr'))
    i+=1
type(html_soup)
```

### Step 3:
#### Create lists to store the data and use pandas to form the list in dataframe.

```python
salary=[]
names=[]
i=0
while i<=10:
    m=0
    while m<=43:
        if rank[i][m].contents[3].text != "SALARY":
            temp =rank[i][m].contents[3].text
            salary.append(temp)
            name=rank[i][m].a.text
            names.append(name)
            m+=1
        else:
            m+=1
    i+=1
playerincome_df=pd.DataFrame({"Player_name":names,"Income":salary})
playerincome_df.head(10)
```

### Step 4:
#### Because we want the top 300 salary of NBA players, we need to sort the data.
#### First we should covert data in serise form str to int.

```python
playerincome_df["Income"]=playerincome_df['Income'].apply(lambda x:x.replace('$',"").replace(',','')).astype('int')
type(playerincome_df['Income'][0])
```
