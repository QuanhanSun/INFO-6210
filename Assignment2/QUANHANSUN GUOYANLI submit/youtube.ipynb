```python
company_channelid=[]
company_channeltitle=[]
company_playlist_title=[]
company_playlist_time=[]
company_description=[]
company_result_name=[]
i = 0
while i <= 8:
    request = youtube.playlists().list(channelId=company_df['channelid'][i],
                                    part='snippet'
                                   )
    temp=request.execute()
    for k in temp['items']:
        company_result_name.append(company_name[i])
        company_channelid.append(k['snippet']['channelId'])
        company_channeltitle.append(k['snippet']['channelTitle'])
        company_playlist_title.append(k['snippet']['title'])
        company_playlist_time.append(k['snippet']['publishedAt'])
    i+=1
company_channelid
company_channeltitle
company_playlist_title
company_playlist_time
```

```python
company_result_df=pd.DataFrame({'company name':company_result_name,
                                'channel id':company_channelid,
                               'channel title':company_channeltitle,
                               'playlist title':company_playlist_title,
                               'post time':company_playlist_time})
company_result_df
```

```python
actors=['Mahershala Ali',
        'Rami Malek',
        'Lady Gaga',
        'Olivia Colman',
        'Yalitza Aparicio',
       ]
```

```python
actor_result_title=[]
actor_result_description=[]
actor_result_posttime=[]
actor_name=[]
i = 0
while i <= 4:
    request = youtube.search().list(q=actors[i],part='snippet',maxResults=20)
    temp=request.execute()
    for k in temp['items']:
        actor_name.append(actors[i])
        actor_result_title.append(k['snippet']['title'])
        actor_result_description.append(k['snippet']['description'])
        actor_result_posttime.append(k['snippet']['publishedAt'])
    i+=1
actor_result_df=pd.DataFrame({'name':actor_name,
                             'video title':actor_result_title,
                             'description':actor_result_description,
                             'post time':actor_result_posttime})
actor_result_df 
```

```python
actor_result_df['post time']=actor_result_df['post time'].apply(lambda x:x.replace('.000Z','').replace('T',' '))
actor_result_df
```

```python
actor_result_df.to_csv('actor_youtube.csv')
```
